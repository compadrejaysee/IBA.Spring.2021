## Checking Stationarity of Data

```python
#Using mean and variance, we can check if there is a shift in the 
#average sales and variance of the divided datasets.
X = dataset['sales'].values
split=int(len(X)/2)
X1,X2=X[0:split],X[split:]
mean1, mean2=X1.mean(),X2.mean()
var1, var2= X1.var(),X2.var()
print('mean1=%f,mean2=%f'%(mean1,mean2))
print('var1=%f,var2=%f'%(var1,var2))
#There is obvious variance in the data since the two datasets have different
#means and variances. 
```

### Using ADFuller Test

```python
#ADFuller Statistic Code
result = adfuller(X)
print('ADF Statistic: %f' % result[0])
if result[0]>1:
    print('Very likely that null hypothesis is true, meaning we have a non-stationary dataset')
elif result[0]<0:
    print('Very likely that null hypothesis is false, meaning we have a stationary dataset')
else:
    print('Hard to say if null hypothesis is true or false, it is complicated')
print('p-value : %f'%result[1])
print(' Critical Values:\n        Key   Value')
for key , value in result[4].items():
    print('\t%s: %.6f' % (key, value))
```

The ADF statistic is greater than the critical value at 10%, meaning we can accept the null hypothesis is true with a significance of 10%. The series is 
### Not Stationary. 
Making the series stationary will allow us to remove randomness from it, and thus allow for better predictions. 

```python
#Using Log Transformation First
result = adfuller(np.log(X))
print('ADF Statistic: %f' % result[0])
if result[0]>1:
    print('Very likely that null hypothesis is true, meaning we have a non-stationary dataset')
elif result[0]<0:
    print('Very likely that null hypothesis is false, meaning we have a stationary dataset')
else:
    print('Hard to say if null hypothesis is true or false, it is complicated')
print('p-value : %f'%result[1])
print(' Critical Values:\n        Key   Value')
for key , value in result[4].items():
    print('\t%s: %.6f' % (key, value))
```

The ADF statistic is greater than the value at 10%, meaning the null hypothesis is true with a significance of 10%; hence it's logical to assume that the series is still non-stationary.

The ADF statistic is greater than the value at 10%, meaning the null hypothesis is true with a significance of 10%; hence it's logical to assume that the series is still non-stationary.

### Trying RMSE Method

```python
X = dataset['sales'].values
window=3
history=[X[i]for i in range(window)]
test = [X[i] for i in range(window, len(X))]
predictions=list()
for t in range (len(test)):
    length=len(history)
    yhat=np.mean([history[i] for i in range(length-window,length)])
    obs=test[t]
    predictions.append(yhat)
    history.append(obs)
    print('predicted=%f, expected=%f' % (yhat, obs))
rmse = (mean_squared_error(test, predictions, squared=False))
print('Test RMSE: %.3f' % rmse)
```

The value of RMSE is 0.02. This shows that our designed model is extremely accurate

## Using AR

```python
# persistence model on training set
train_pred = [x for x in train_X]
# calculate residuals
train_resid = [train_y[i]-train_pred[i] for i in range(len(train_pred))]
# model the training set residuals
model = AR(train_resid)
model_fit = model.fit()
window = model_fit.k_ar
prediction = model_fit.forecast(steps=5)
coef = model_fit.params
print('Lag=%d \nCoef=%s' % (window, coef))
```

```python
history = train_resid[len(train_resid)-window:]
history = [history[i] for i in range(len(history))]
```

```python
train_resid
```

```python
# walk forward over time steps in test
history = train_resid[len(train_resid)-window:]
history = [history[i] for i in range(len(history))]
predictions = list()
expected_error = list()
for t in range(len(test_y)):
# persistence
    yhat = test_X[t]
    error = test_y[t] - yhat
    expected_error.append(error)
# predict error
    length = len(history)
    lag = [history[i] for i in range(length-window,length)]
    pred_error = coef[0]
    for d in range(window):
        pred_error += coef[d+1] * lag[window-d-1]
    predictions.append(pred_error)
    history.append(error)
    print('predicted error=%f, expected error=%f' % (pred_error, error))
# plot predicted error
plt.plot(expected_error)
plt.plot(predictions, color='red')
plt.show()
```

```python
# walk forward over time steps in test
history = train_resid[len(train_resid)-window:]
history = [history[i] for i in range(len(history))]
predictions = list()
expected_error = list()
for t in range(len(test_y)):
# persistence
    yhat = test_X[t]
    error = test_y[t] - yhat
    expected_error.append(error)
# predict error
    length = len(history)
    lag = [history[i] for i in range(length-window,length)]
    pred_error = coef[0]
    for d in range(window):
        pred_error += coef[d+1] * lag[window-d-1]
    # correct the prediction
    yhat = yhat + pred_error
    predictions.append(yhat)
    history.append(error)
    print('predicted error=%f, expected error=%f' % (pred_error, error))
rmse = (mean_squared_error(test_y, predictions,squared=False))
print('Test RMSE: %.3f' % rmse)
# plot predicted error
plt.plot(test_y)
plt.plot(predictions, color='red')
plt.show()
```
