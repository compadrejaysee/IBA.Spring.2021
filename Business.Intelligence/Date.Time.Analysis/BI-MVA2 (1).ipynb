```python
#importing the basic libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plot
import missingno as mano
%matplotlib inline
```

```python
#reading the CSV file
missingdf = pd.read_csv("Missing.Value.Data.csv")
```

```python
#Performing imputation - imputation is the process of replacing missing values either thorugh mean, median or mode
from sklearn.impute import SimpleImputer

#creating a separate data frame for imputation
missingdf_mean = missingdf.copy(deep=True)

mean_imputation = SimpleImputer(strategy='mean') #Replacing missing values with mean

#take only columns where mean imputation matters, i.e., numerical columns
missingdf_mean[['Customer.Age','Sales','Quantity']] = mean_imputation.fit_transform(missingdf_mean[['Customer.Age','Sales','Quantity']])
```

```python
missingdf
```

```python
missingdf_mean
```

```python
#checking if there are any missing values still present in the imputated data
missingdf_mean.isnull().sum()
```

```python
#No missing data is found in numerical attributes 
#comparing sales vs quantity (values filled in by mean) through scatter plot
missingdf_mean.plot(x='Sales',y='Quantity',kind='scatter',alpha=0.5,cmap='rainbow')
```

```python
#determining those rows where values of either sales or quantity went missing and incorporating them in the above scatter plot 
nulls =  missingdf['Sales'].isnull() + missingdf['Quantity'].isnull()
missingdf_mean.plot(x='Sales',y='Quantity',kind='scatter',alpha=0.5,c=nulls,cmap='rainbow')
```

```python
#Since blue marks in the first scatter plot is following the same pattern as red marks in second scatter plot, this shows that
#missing values are filled in accordance to the given distribution
```

```python
#comparing sales vs customer age (values filled in by mean) through scatter plot
missingdf_mean.plot(x='Sales',y='Customer.Age',kind='scatter',alpha=0.5,cmap='rainbow')
```

```python
#determining those rows where values of either sales or customer age went missing and incorporating them in the above scatter plot 
nulls =  missingdf['Sales'].isnull() + missingdf['Customer.Age'].isnull()
missingdf_mean.plot(x='Sales',y='Customer.Age',kind='scatter',alpha=0.5,c=nulls,cmap='rainbow')
```

```python
#Since blue marks in the third scatter plot is following the same pattern as red marks in fourth scatter plot, this shows that
#missing values are filled in accordance to the given distribution
```

```python
#read the CSV file again
missingdf = pd.read_csv("Missing.Value.Data.csv")
```

```python
#Performing imputation thorugh K-Nearest Neighbour strategy
from sklearn.impute import KNNImputer

#missing values are imputed using the mean value from n_neighbors=2 i.e. nearest neighbors found in the data.
imputer = KNNImputer (n_neighbors=2)

#Creating a seperate data frame for imputation
missingdf_knn = missingdf.copy(deep=True)
missingdf_knn[['Customer.Age','Sales','Quantity']] = imputer.fit_transform(missingdf_knn[['Customer.Age','Sales','Quantity']])
missingdf_knn
```

```python
#comparing sales vs quantity (values filled in by knn)
missingdf_knn.plot(x='Sales',y='Quantity',kind='scatter',alpha=0.5,cmap='rainbow')
```

```python
#determining those rows where values of either sales or quantity went missing and incorporating them in the above scatter plot 
nulls =  missingdf['Sales'].isnull() + missingdf['Quantity'].isnull()
missingdf_knn.plot(x='Sales',y='Quantity',kind='scatter',alpha=0.5,c=nulls,cmap='rainbow')
```
